<h1> Gaussian Mixture Models - Notebook </h1>
The results of the Gaussian Mixture models clustering are recorded in this notebook. We tried also PCA with GME, but did not see any added value by using GME compared to kmeans or DBSCAN. Thats why we stopped working on GME.

```python
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("data\\analysis_dataset.csv", sep=";")

del data["Unnamed: 0"]
del data["coin"]
del data["axis"]
del data["mindiam"]
del data["findspot"]
```

```python
imp = IterativeImputer(max_iter=10, random_state=0)
imp_data = data.copy()
head = data.columns
for col in data.columns:
    imp.fit(np.array(data[col]).reshape(-1,1))
    imp_data[col] = imp.transform(np.array(data[col]).reshape(-1,1))
```

Get best parameters for GME with original data

```python
n_range = range(2,11)

# Create empty lists to store the BIC and AIC values
bic_score = []
aic_score = []
X = imp_data.copy()
# Loop through the range and fit a model
for n in n_range:
    gm = GaussianMixture(n_components=n, 
                         random_state=123, 
                         n_init=10)
    gm.fit(X)
    
    # Append the BIC and AIC to the respective lists
    bic_score.append(gm.bic(X))
    aic_score.append(gm.aic(X))
    
# Plot the BIC and AIC values together
fig, ax = plt.subplots(figsize=(12,8),nrows=1)
ax.plot(n_range, bic_score, '-o', color='orange')
ax.plot(n_range, aic_score, '-o', color='green')
ax.set(xlabel='Number of Clusters', ylabel='Score')
ax.set_xticks(n_range)
ax.set_title('BIC and AIC Scores Per Number Of Clusters')
```

Same for PCA

```python
X = imp_data.copy()
n_comp = range(1,11)
for i in n_comp:
    pca = PCA(n_components=i)
    pca.fit(X)
    X_pca = pca.transform(X)

    n_range = range(2,11)

    # Create empty lists to store the BIC and AIC values
    bic_score = []
    aic_score = []
    
    # Loop through the range and fit a model
    print(f"{i}-PCA transformation")
    for n in n_range:
        gm = GaussianMixture(n_components=n, 
                             random_state=123, 
                             n_init=10)
        gm.fit(X_pca)

        # Append the BIC and AIC to the respective lists
        bic_score.append(gm.bic(X_pca))
        aic_score.append(gm.aic(X_pca))

    # Plot the BIC and AIC values together
    fig, ax = plt.subplots(figsize=(12,8),nrows=1)
    ax.plot(n_range, bic_score, '-o', color='orange')
    ax.plot(n_range, aic_score, '-o', color='green')
    ax.set(xlabel='Number of Clusters', ylabel='Score')
    ax.set_xticks(n_range)
    ax.set_title('BIC and AIC Scores Per Number Of Clusters')
    plt.show()
```

select pca-> 8 features and 6 clusters

```python
pca = PCA(n_components=8)
pca.fit(imp_data)
X_pca = pca.transform(imp_data)

gm = GaussianMixture(n_components=6, 
                     random_state=123, 
                     n_init=10)
gm.fit(X_pca)
results = gm.predict(X_pca)
```

```python
data["labels"] = results
```
