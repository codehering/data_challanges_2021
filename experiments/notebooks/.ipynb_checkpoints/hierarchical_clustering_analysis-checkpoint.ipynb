<h1> Hierarchical Clustering Analysis Notebook </h1>
The results of the hierarchical clustering without dimension reduction are recorded in this notebook. 
You can see from the individual plots below that the results are very difficult to interpret. 
The results of the dbscan clustering are recorded in this notebook. First of all, a dendrogram analysis was carried out, and the greatest explanatory power of the clusters was detected at a size of 3 (longest distance in the dendrogram without a horizontal line).  It can be seen that the clusters are very difficult to identify and interpret without dimension reduction. 


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as shc
```

```python

data = pd.read_csv('data\\analysis_dataset_w_material.csv', sep=";")
#del data["mindiam"]
del data["axis"]
del data["Unnamed: 0"]
del data["findspot"]
del data["material"]
del data["collection_cat"]
del data["coin"]
del data["mindiam"]
del data["maxdiam"]
del data["startdate"]
data = data.dropna()
data.shape
data.columns
```

```python
plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
cluster.fit_predict(data)
```

```python
data_np =np.array(data)
plt.figure(figsize=(10, 7))
plt.scatter(data_np[:,0], data_np[:,1], c=cluster.labels_, cmap='rainbow')
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data_np[:,0], data_np[:,2], c=cluster.labels_, cmap='rainbow')
```

```python
data["cluster"] = cluster.labels_
cluster1 = data[data["cluster"]==0]
cluster2 = data[data["cluster"]==1]
cluster3 = data[data["cluster"]==2]
```
