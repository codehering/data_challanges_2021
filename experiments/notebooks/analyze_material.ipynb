```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/analysis_dataset_w_material.csv", sep=";")
```

```python
data.columns
```

```python
subset_data = data[["enddate", "material"]]
subset_data = subset_data.dropna()
subset_data.shape
```

```python
subset_data["enddate_round"] = (subset_data["enddate"] // 10)*10
```

```python
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100
subset_data.pivot_table(index='enddate_round',columns='material',values='enddate',aggfunc='count').plot()
plt.xlabel("Enddate")
plt.ylabel("Frequency")
plt.show()
```
